````python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

df = pd.read_csv("college_completion.csv")

# dimensions
print("Data Dimensions:", df.shape)
print("Number of Observations", df.shape[0])
print("Variables:", df.columns.tolist())
print("\nFirst few rows:\n", df.head())

# cross tabulation of Control and Level
cross_tab = pd.crosstab(df["control"], df["level"])
print("\nCross Tabulation of Control and Level:\n", cross_tab)
print("""
Patterns Observed:
- public institutions are heavily concentrated in 2-year colleges
- Private for-profit institutions are also mainly 4-year, but persent to a lesser extent in 2-year
- Private not-for-profit institutions are mainly 4-year
""")

# grad 100 analysis + histogram + kernel density plot + boxplot
print("\nStatistical Description of grad_100_value:\n", df["grad_100_value"].describe())

plt.figure(figsize=(12,6))
plt.subplot(2,2,1)
sns.histplot(df["grad_100_value"].dropna(), kde=False)
plt.title("Histogram of grad_100_value")

plt.subplot(2,2,2)
sns.kdeplot(df["grad_100_value"].dropna())
plt.title("Kernel Density Plot of grad_100_value")

plt.subplot(2,2,3)
sns.boxplot(x=df["grad_100_value"].dropna())
plt.title("Boxplot of grad_100_value")

plt.tight_layout()
plt.show()

# group kernel density plot
plt.figure(figsize=(14,6))
plt.subplot(1,2,1)
for control_type in df["control"].unique():
    sns.kdeplot(df[df["control"] == control_type]["grad_100_value"].dropna(), label=control_type)
plt.title("Groupe Kernel Density Plot by Control")
plt.xlabel("grad_100_value")
plt.legend()

plt.subplot(1,2,2)
for level_type in df["level"].unique():
    sns.kdeplot(df[df["level"] == level_type]["grad_100_value"].dropna(), label=level_type)
plt.title("Groupe Kernel Density Plot by Level")
plt.xlabel("grad_100_value")
plt.legend()

plt.tight_layout()
plt.show()

print("\nGrouped statistical descriptions of grad_100_value by level:\n", df.groupby("level")["grad_100_value"].describe())
print("\nGrouped statistical descriptions of grad_100_value by control:\n", df.groupby("control")["grad_100_value"].describe())
print("""
Observations:
- 4 year institutions generally have higher graduation rates than 2-year institutions
- Public and private not-for-profit institutions have higher graduation rates than private for-profit institutions
""")

# interaction variable
df["levelXcontrol"] = df["level"] + ", " + df["control"]

plt.figure(figsize=(12,6))
for combo in df["levelXcontrol"].unique():
    sns.kdeplot(df[df["levelXcontrol"] == combo]["grad_100_value"].dropna(), label=combo)
plt.title("Groupe Kernel Density Plot by Level and Control interaction")
plt.xlabel("grad_100_value")
plt.legend()
plt.show()
print("""
Observations:
- Institutions with the best graduation rates are likely: 4-year, private not-for-profit, as their KDE is shifted furthest to the right
""")

# aid value
print("\nStatistical Description of aid_value:\n", df["aid_value"].describe())

plt.figure(figsize=(12,6))
sns.kdeplot(df["aid_value"].dropna())
plt.title("Kernel Density Plot of aid_value")
plt.show()

plt.figure(figsize=(14,6))
plt.subplot(1,2,1)
for control_type in df["control"].unique():
    sns.kdeplot(df[df["control"] == control_type]["aid_value"].dropna(), label=control_type)
plt.title("Groupe Kernel Density Plot of aid_value by Control")
plt.xlabel("aid_value")
plt.legend()

plt.subplot(1,2,2)
for level_type in df["level"].unique():
    sns.kdeplot(df[df["level"] == level_type]["aid_value"].dropna(), label=level_type)
plt.title("Groupe Kernel Density Plot of aid_value by Level")
plt.xlabel("aid_value")
plt.legend()

plt.tight_layout()
plt.show()

print("\nGrouped statistical descriptions of aid_value by level:\n", df.groupby("level")["aid_value"].describe())
print("\nGrouped statistical descriptions of aid_value by control:\n", df.groupby("control")["aid_value"].describe())
print("""
Observations:
- The shape of the graph is most likely explain by:
* The bimodal nature that results from different funding models between instituion types (public vs. private)
* Government aid impacts aid averages as does institutional endowment
""")

# Scatteplots of grad_100_value vs aid_value
plt.figure(figsize=(8,6))
sns.scatterplot(x=df["aid_value"],y=df["grad_100_value"],alpha=0.5)
plt.title("Scatterplot of grad_100_value vs aid_value")
plt.xlabel("aid_value")
plt.ylabel("grad_100_value")
plt.show()

plt.figure(figsize=(14,6))
plt.subplot(1,2,1)
sns.scatterplot(x="aid_value", y="grad_100_value", hue="level", data=df, alpha=0.7)
plt.title("Scatterplot of grad_100_value vs aid_value by level")
plt.xlabel("aid_value")
plt.ylabel("grad_100_value")

plt.subplot(1,2,2)
sns.scatterplot(x="aid_value", y="grad_100_value", hue="control", data=df, alpha=0.7)
plt.title("Scatterplot of grad_100_value vs aid_value by control")
plt.xlabel("aid_value")
plt.ylabel("grad_100_value")

plt.tight_layout()
plt.show()

print("""
Observations:
- There is a weak positive correlation overall
- For 4-year institutions, there appears to be a slightly more postive trend
- aid seems to have a more pronounced positive effect on graduation rates in 4-year institutions and private institutions, specifically private not-for-profit institutions
- For public institutions, there's less of a clear relationship compared to private institutions
""")
````
